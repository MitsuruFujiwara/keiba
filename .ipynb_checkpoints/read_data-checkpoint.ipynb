目的値の変換。
１位の場合は1、それ以外は0となるように変換し、２値分類問題に帰着させる。

```python
def y(result):
    for t in result:
        if t=='1':
            yield 1
        else:
            yield 0

race_result['y'] = list(y(race_result['order_of_finish']))
```

finishing_timeの変換。秒単位にする。

```python
def _time(finishing_time):
    for ft in finishing_time:
        _ft = ft.split(':')
        try:
            yield float(_ft[0])*60 + float(_ft[1])
        except ValueError:
            yield np.nan
            
race_result['time'] = list(_time(race_result['finishing_time']))
```
